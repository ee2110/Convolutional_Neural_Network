```
import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
img = cv2.imread('color_palette.jpg')
img=cv2.resize(img,(700,500))
img
```

```
# plot image
def plotImages(image_arr):
  fig, axes = plt.subplots(1, 5, figsize=(20,20))
  axes = axes.flatten()
  for img, ax in zip( image_arr, axes):
    ax.imshow(img)
  plt.tight_layout()
  plt.show()
```

```
plotImages(img)
```
